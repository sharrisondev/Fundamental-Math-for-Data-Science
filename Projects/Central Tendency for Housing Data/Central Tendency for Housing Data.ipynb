# NYC Housing Data – Mean, Median, Mode
In this notebook, we compare rental prices for one-bedroom apartments in Brooklyn, Manhattan, and Queens.  
We’ll calculate the **mean, median, and mode** for each borough using NumPy, Pandas, and SciPy.

```python
# Import packages
import numpy as np
import pandas as pd
from scipy import stats
```

```python
# Read in housing data
brooklyn_one_bed = pd.read_csv('brooklyn-one-bed.csv')
manhattan_one_bed = pd.read_csv('manhattan-one-bed.csv')
queens_one_bed = pd.read_csv('queens-one-bed.csv')

# Extract the rent columns
brooklyn_price = brooklyn_one_bed['rent']
manhattan_price = manhattan_one_bed['rent']
queens_price = queens_one_bed['rent']
```

```python
# Mean
brooklyn_mean = np.average(brooklyn_price)
manhattan_mean = np.average(manhattan_price)
queens_mean = np.average(queens_price)

# Median
brooklyn_median = np.median(brooklyn_price)
manhattan_median = np.median(manhattan_price)
queens_median = np.median(queens_price)

# Mode
brooklyn_mode = stats.mode(brooklyn_price, keepdims=True)
manhattan_mode = stats.mode(manhattan_price, keepdims=True)
queens_mode = stats.mode(queens_price, keepdims=True)
```

```python
print(f"Brooklyn – Mean: {brooklyn_mean:.2f}, Median: {brooklyn_median}, Mode: {brooklyn_mode[0][0]} ({brooklyn_mode[1][0]} times)")
print(f"Manhattan – Mean: {manhattan_mean:.2f}, Median: {manhattan_median}, Mode: {manhattan_mode[0][0]} ({manhattan_mode[1][0]} times)")
print(f"Queens – Mean: {queens_mean:.2f}, Median: {queens_median}, Mode: {queens_mode[0][0]} ({queens_mode[1][0]} times)")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(brooklyn_price, bins=30, alpha=0.5, label="Brooklyn")
plt.hist(manhattan_price, bins=30, alpha=0.5, label="Manhattan")
plt.hist(queens_price, bins=30, alpha=0.5, label="Queens")
plt.legend()
plt.xlabel("Rent ($)")
plt.ylabel("Frequency")
plt.title("Distribution of One-Bedroom Rental Prices in NYC")
plt.show()
```
